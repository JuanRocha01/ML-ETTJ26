## 1. Introdução
A interpolação flat-forward é uma metodologia de construção de curvas de juros na qual se assume que a taxa forward instantânea permanece constante entre dois vértices consecutivos da curva. No framework desenvolvido neste estudo, ela será utilizada como o modelo naïve de referência: uma abordagem determinística, local e de baixa complexidade, contra a qual modelos paramétricos e não paramétricos mais sofisticados poderão ser comparados, inclusive comparar o uso puro da interpolação com o bootstrapping.

É importante distinguir a interpolação flat-forward do processo de bootstrapping propriamente dito. O bootstrapping é o procedimento utilizado para extrair fatores de desconto ou taxas zero a partir dos preços observados dos instrumentos financeiros. Já o flat-forward define como a curva se comporta entre os vértices conhecidos. Assim, depois que cada instrumento fornece informação sobre um ou mais pontos da estrutura a termo, a hipótese flat-forward permite preencher os intervalos intermediários de maneira consistente com a precificação por ausência de arbitragem.

## Monotonicidade dos Fatores de Desconto
Uma curva de juros livre de arbitragem deve representar de maneira consistente o valor temporal do dinheiro. Seja $P(t,T)$ o fator de desconto, na data $t$, associado ao pagamento de uma unidade monetária no vencimento $T$. Equivalentemente, $P(t,T)$ pode ser interpretado como o preço, na data $t$, de um título zero cupom livre de risco que paga uma unidade monetária em $T$.

Para dois vencimentos $T_i < T_{i+1}$, a condição de monotonicidade é
$$P(t,T_{i+1})\leq P(t,T_i)$$
Sob taxas nominais não negativas, um pagamento realizado em uma data mais distante não pode valer mais, na data presente do que o mesmo pagamento realizado em uma data anterior. Assim, os fatores de desconto devem ser positivos e não crescentes em relação ao vencimento:
$$1=P(t,t) \geq P(t,T_1) \geq P(t,T_2) \geq \cdots > 0$$

### Condições de Arbitragem 
Considere dois títulos zero cupom com vencimentos $T_1$e $T_2$, sendo $T_1<T_2$.Os respectivos preços são $P(t,T_1)$ e $P(t,T_2)$.

Suponha que a curva estimada viole a monotonicidade dos fatores de desconto, isto é,
$P(t,T_2)>P(t,T_1)$.Nesse caso, o título que paga uma unidade monetária em uma data mais distante possui preço superior ao título que paga a mesma unidade em uma data anterior.

Pode-se construir a seguinte estratégia na data $t$:
- 1. vender a descoberto uma unidade do título com vencimento em $T_2$, recebendo $P(t,T_2)$

- 2. comprar uma unidade do título com vencimento em $T_1$, pagando $P(t,T_1)$. 

O fluxo de caixa líquido inicial da operação é
$$\Pi_t=P(t,T_2)-P(t,T_1)$$
Como foi assumido que $P(t,T_2)>P(t,T_1)$, tem-se $\Pi_t>0$.

No vencimento $T_1$, o título comprado paga uma unidade monetária:
$$CF_{T_1}=+1$$
Essa unidade pode ser mantida em caixa entre (T_1) e (T_2). Em (T_2), ela é utilizada para liquidar a obrigação decorrente da venda a descoberto do título mais longo:

$$CF_{T_2}=1-1=0$$

Portanto, a estratégia apresenta:

$\Pi_t>0$

e nenhum fluxo líquido negativo nas datas futuras. Trata-se, assim, de uma oportunidade de arbitragem.

O ganho de arbitragem, medido na data inicial, é diretamente dado por
$$P(t,T_2)-P(t,T_1)$$
Logo, sempre que
$$A(t;T_1,T_2)>0,$$
a curva apresenta uma violação da monotonicidade dos fatores de desconto.

Essa demonstração pressupõe que uma unidade monetária recebida em (T_1) pode ser mantida até (T_2) sem perda nominal. Em mercados nos quais taxas negativas são admissíveis ou em que o armazenamento de moeda possui custos relevantes, a interpretação precisa ser feita com maior cuidado. Para o mercado brasileiro de juros prefixados, no qual o regime usual é de taxas nominais positivas, essa condição constitui um teste econômico apropriado para a curva.
